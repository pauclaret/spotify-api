```python
import pyjsonviewer
import sys
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials
import pandas as pd
import numpy as np
import os
from dotenv import load_dotenv
load_dotenv()
```

```python
client_secret = os.getenv("secret")
client_id = os.getenv("id")
```

```python
credenciales = SpotifyClientCredentials(client_id, client_secret)
sp = spotipy.Spotify(client_credentials_manager=credenciales)
```

```python
url = "https://open.spotify.com/playlist/6sAe071wJ5aJF7CJUgo0Me?si=4efc4513a0f44539"
```

```python
offset = 0
resultados = []
for i in range(3):
    resultados.append(sp.playlist_tracks(uri, offset = offset)["items"])
    offset += 100
```

```python
diccionario_spot = {"uri_cancion": [],
                    "nombre_cancion": [],
                    "fecha_incorporacion": [],
                    "fecha_publicacion": [],
                    "explicita": [],
                    "ironhacker": [],
                    "artistas": []}
```

```python
for i in range(len(resultados)):
    for z in range(len(resultados[i])):
        diccionario_spot["uri_cancion"].append(resultados[i][z]["track"]["uri"])
        diccionario_spot["nombre_cancion"].append(resultados[i][z]["track"]["name"])
        diccionario_spot["fecha_incorporacion"].append(resultados[i][z]["added_at"])
        diccionario_spot["fecha_publicacion"].append(resultados[i][z]["track"]["album"]["release_date"])
        diccionario_spot["explicita"].append(resultados[i][z]["track"]["explicit"])
        diccionario_spot["ironhacker"].append(resultados[i][z]["added_by"]["id"])

        artistas = resultados[i][z]["track"]["artists"]
        nombre_artistas = []

        for artista in range(len(artistas)):
            nombre_artistas.append(artistas[artista]["name"])
        diccionario_spot["artistas"].append(nombre_artistas)

        
```

```python
for k, v in diccionario_spot.items():
    print(k, len(v))
```

```python
df = pd.DataFrame(diccionario_spot)
df.head()
```

```python
uris_canciones = df["uri_cancion"].unique().tolist()
uris_canciones[:5]
```

```python
features = []

for i in uris_canciones:
    features.append(sp.audio_features(i))
```

```python
features[0]
```

```python
df_features = pd.DataFrame(features)
df_features.head()
```

```python
df_features = df_features[0].apply(pd.Series)
df_features.head()
```

```python
df_final = df.merge(df_features, left_on = "uri_cancion", right_on = "uri")
df_final.head()
```
